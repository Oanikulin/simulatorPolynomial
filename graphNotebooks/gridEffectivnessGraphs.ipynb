```python
data = open("gridEffectivnessData.txt", "r")
```

```python
gridFullTime = [float(line.strip().split()[2]) for line in data][::2]
data.seek(0, 0)
grifHalfTime = [float(line.strip().split()[2]) for line in data][1::2]
data.seek(0, 0)
gridFullCount = [float(line.strip().split()[3]) for line in data][::2]
data.seek(0, 0)
grifHalfCount = [float(line.strip().split()[3]) for line in data][1::2]
data.seek(0, 0)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(15, 20))
ax = fig.add_subplot(2, 1, 1)

yvalues = list(range(3, 26))
line, = ax.plot(yvalues, gridFullTime[:len(yvalues)], color='blue', lw=2, label="full iteration")

line2, = ax.plot(yvalues, grifHalfTime[:len(yvalues)], color='red', lw=2, label="half iteration")

gridTmp = grifHalfTime[3::2]

line3, = ax.plot(yvalues[:len(yvalues)], gridTmp, color='green', lw=2, label="half iteration (shifted)")

ax.legend(loc="upper left", prop={'size': 25})
plt.xlabel("Depth of graph")
plt.ylabel("Time (mcs)")

ax.set_yscale('log')
plt.show()
```

```python
fig = plt.figure(figsize=(15, 20))
ax = fig.add_subplot(2, 1, 1)

yvalues = list(range(3, 26))
line, = ax.plot(yvalues, gridFullCount[:len(yvalues)], color='blue', lw=2, label="full iteration")

line2, = ax.plot(yvalues, grifHalfCount[:len(yvalues)], color='red', lw=2, label="half iteration")

gridTmp = grifHalfCount[3::2]

line3, = ax.plot(yvalues[:len(yvalues)], gridTmp, color='green', lw=2, label="half iteration (shifted)")

ax.legend(loc="upper left", prop={'size': 25})
plt.xlabel("Depth of graph")
plt.ylabel("Number of points")

ax.set_yscale('log')
plt.show()
```

```python
fig = plt.figure(figsize=(15, 20))
ax = fig.add_subplot(2, 1, 1)

yvalues = list(range(4, 26))
xvalues = [gridFullTime[i] / gridFullTime[i - 1] for i in range(1, 23)]
line, = ax.plot(yvalues, xvalues, color='blue', lw=2, label="growth")

ax.legend(loc="upper left", prop={'size': 25})
plt.xlabel("Depth of graph")
plt.ylabel("Time proportion")

plt.show()
```
