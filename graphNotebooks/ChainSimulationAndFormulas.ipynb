```python
data = open("chainIterationData.txt", "r")
```

```python
theoreticalData = [float(line.strip().split()[1]) for line in data][:30]
data.seek(0, 0)
simData = [float(line.strip().split()[1]) for line in data][30:]
data.seek(0, 0)
simTime = [float(line.strip().split()[0]) for line in data][30:]
data.seek(0, 0)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(15, 20))
ax = fig.add_subplot(2, 1, 1)

yvalues = list(range(1, 31))
line, = ax.plot(yvalues, theoreticalData, "o", color='red', lw=2, label="theoretical formula")

line2, = ax.plot(simTime, simData, color='blue', lw=2, label="simulation")

ax.legend(loc="upper left", prop={'size': 25})
plt.xlabel("Depth of graph")
plt.ylabel("Number of points")

ax.set_yscale('log')
plt.show()
```
