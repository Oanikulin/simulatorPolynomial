```python
chainData = open("chainData.txt", "r")
treeData = open("randomTreeData.txt", "r")
fullData = open("fullTreeData.txt", "r")
```

```python
chainFullTime = [float(line.strip().split()[2]) for line in chainData][::2]
chainData.seek(0, 0)
chainHalfTime = [float(line.strip().split()[2]) for line in chainData][1::2]
chainData.seek(0, 0)
chainFullCount = [float(line.strip().split()[3]) for line in chainData][::2]
chainData.seek(0, 0)
chainHalfCount = [float(line.strip().split()[3]) for line in chainData][1::2]
chainData.seek(0, 0)
```

```python
randomFullTime = [float(line.strip().split()[2]) for line in treeData][::2]
treeData.seek(0, 0)
randomHalfTime = [float(line.strip().split()[2]) for line in treeData][1::2]
treeData.seek(0, 0)
randomFullCount = [float(line.strip().split()[3]) for line in treeData][::2]
treeData.seek(0, 0)
randomHalfCount = [float(line.strip().split()[3]) for line in treeData][1::2]
treeData.seek(0, 0)
```

```python
fullFullTime = [float(line.strip().split()[2]) for line in fullData][::2]
fullData.seek(0, 0)
fullHalfTime = [float(line.strip().split()[2]) for line in fullData][1::2]
fullData.seek(0, 0)
fullFullCount = [float(line.strip().split()[3]) for line in fullData][::2]
fullData.seek(0, 0)
fullHalfCount = [float(line.strip().split()[3]) for line in fullData][1::2]
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(15, 20))
ax = fig.add_subplot(2, 1, 1)

yvalues = list(range(3, 36))
line, = ax.plot(yvalues, chainFullCount, color='blue', lw=2, label="full iteration")

line2, = ax.plot(yvalues, chainHalfCount, color='red', lw=2, label="half iteration")

chainTmp = chainHalfCount[4::2]

line3, = ax.plot(yvalues[:15], chainTmp, color='green', lw=2, label="half iteration (shifted)")

ax.legend(loc="upper left", prop={'size': 25})
plt.xlabel("Depth of graph")
plt.ylabel("Number of points")

ax.set_yscale('log')
plt.show()
```
